## Flow-chart etude NIDA

```python
# Import libraries
import pandas as pd

# Read data
dm01 = pd.read_csv("dm1.csv")
dm02 = pd.read_csv("dm1.csv")

print(dm01['ARM'].value_counts())
print(dm02['ARM'].value_counts())

dm = pd.concat([dm01, dm02], ignore_index=True)

# Verify unique patients
print((dm['USUBJID'].value_counts() > 1).sum())

# Description of number of randomized patients
print(dm['ARM'].value_counts())

dmIncluded = dm[dm['ARM'] != "SCREEN FAILURE"]
print(dmIncluded['ARM'].value_counts())

# Discharge
ds01 = pd.read_csv("ds1.csv")
ds02 = pd.read_csv("ds2.csv")

ds = pd.concat([ds01, ds02], ignore_index=True)

dsDischarged = (ds[ds['EPOCH'] != "SCREENING"]
                .assign(beforeEndActive=lambda x: x['VISITNUM'] < 14)
                .query('beforeEndActive & DSDECOD != "PARTICIPANT COMPLETED ACTIVE PHASE OF STUDY"')
                [['STUDYID', 'USUBJID', 'DSDECOD', 'VISITNUM']]
                .drop_duplicates())

print(dsDischarged['VISITNUM'].value_counts())
print(dsDischarged['DSDECOD'].value_counts())

dtf = dmIncluded.merge(dsDischarged, on='USUBJID', how='left')

# Cross-tabulation with row sums
crosstab = pd.crosstab(dtf['ARMCD'], dtf['VISITNUM_y'])
print(crosstab.sum(axis=1))

print(pd.crosstab(dtf['DSDECOD'], dtf['ARMCD']))
```
